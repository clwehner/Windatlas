```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Weibull(
    windspeed: np.array,
    k: float,
    A: float):
    """
    """
    return k/A*(windspeed/A)**(k-1)*np.exp(-(windspeed/A)**k)

def WeibullIntegral(
    windspeed: np.array,
    k: float,
    A: float):
    """
    """
    return -np.exp(-(windspeed**k)/(A**k))

def windspeedArray(
    increment:float,
    start:float = 0.,
    end:float = 26.):
    """
    """
    num = int((end-start)/increment) + 1
    return np.linspace(0.,26.,num=num)
```

```python
# Windspeed generater
ws = windspeedArray(increment=0.0001)

# Weibull generator
wb = Weibull(windspeed=ws, k=2., A=6.) * 100

plt.plot(ws, wb)
```
